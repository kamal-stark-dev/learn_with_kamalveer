A list of messages can be provided to a chat model conversation history and each message has a role that models use to indicate who send the message in the conversation.

````python
# Dictionary Format
conversation = [
    {"role": "system", "content": "You are a helpful assistant that translates English to Hindi."},
    {"role": "user", "content": "Translate: I love programming."},
    {"role": "assistant", "content": "मुझे प्रोग्रामिंग पसंद है |"},
    {"role": "user", "content": "Translate: LangChain are really cool."}
]

response = llm.invoke(conversation)
print(response.content)
````

````python
# Message Objects
from langchain.messages import HumanMessage, AIMessage, SystemMessage

conversation = [
    SystemMessage("You are a helpful assistant that tells the genre of a movie or a show."),
    HumanMessage("What's the genre of Avengers: Endgame"),
    AIMessage("Avengers Endgame is officially classified as a *superhero film*, it's a blend of *Action*, *Adventure* and *Sci-Fi*."),
    HumanMessage("Ratatouille"),
    AIMessage("Ratatouille is an *animated comedy-drama*."),
    HumanMessage("Zankano Tachi"),
    AIMessage("Can't find information about the movie."),
    HumanMessage("Attack on Titan")
]

response = llm.invoke(conversation)
print(response.content)
````

So you can send requests in the following ways:

### Stream
Most models can stream their output content while it is being generated. By displaying output progressively, streaming significantly improves user experience, particularly for longer responses.

As we can see the response above took 12 seconds to generate, so streaming here can make the user's experience better.

````python
# basic text streaming
for chunk in llm.stream("Why do parrots have colorful feathers?"):
    print(chunk.text, end="|", flush=True)
````

````python
# stream tool calls, reasoning, and other content

for chunk in llm.stream("What color is the sky?"):
    for block in chunk.content_blocks:
        if block["type"] == "reasoning" and (reasoning := block.get("reasoning")):
            print(f"Reasoning: {reasoning}")
        elif block["type"] == "tool_calling_chunk":
            print(f"Tool call chunk: {block}")
        elif block["type"] == "text":
            print(block["text"])
        else:
            pass
````
